```python
import pandas as pd
import numpy as np
import os
```

```python
state = {
    "run":0,
    "sit":1,
    "walk":2
}
```

```python

def pad_columns_with_zeros(data, target_cols):
    if data.shape[1] < target_cols:
        diff = target_cols - data.shape[1]
        padding = np.zeros((data.shape[0], diff))
        data = np.hstack((data, padding))

    return data
```

```python
file_num=0
signals = None
Y = []
data_folder = os.path.join(os.path.dirname(os.getcwd()), 'non-arrhythmia', "dataset", "csv")

for filename in os.listdir(data_folder):
    signal_file = pd.read_csv(os.path.join(data_folder, filename))

    # Przetwarzanie każdej kolumny indywidualnie
    for col in ["pleth_4", "pleth_5", "pleth_6"]:
        # Jeśli signals jest None, inicjalizujemy go pierwszym plikiem
        if signals is None:
            signals = signal_file[col].values.reshape(1, -1)
        else:
            # Znalezienie maksymalnej liczby kolumn
            max_cols = max(signals.shape[1], signal_file[col].shape[0])

            # Wypełnianie kolumn zerami w tablicy numpy
            signals = pad_columns_with_zeros(signals, max_cols)
            new_signal = pad_columns_with_zeros(signal_file[col].values.reshape(1, -1), max_cols)

            # Dodawanie nowego wiersza do signals
            signals = np.vstack((signals, new_signal))

    base_state = state[os.path.splitext(filename)[0].split("_")[1]]
    Y.append((0, base_state, 0))
    Y.append((0, base_state, 1))
    Y.append((0, base_state, 2))

    print(file_num)
    file_num+=1

print("Signals shape:", signals.shape)
# print("Labels (Y):", Y)
```

```python
signals=pd.DataFrame(signals)
```

```python
import pandas as pd
import numpy as np

# Zakładamy, że df to Twój DataFrame o wymiarach (198, 275656)
original_df = pd.DataFrame(np.random.rand(198, 275656))  # Przykładowy DataFrame

# Określamy nową liczbę kolumn
target_cols = 30000

# Funkcja do dzielenia wiersza na fragmenty
def split_row(row, chunk_size):
    return [row[i:i + chunk_size] for i in range(0, len(row), chunk_size)]

# Tworzymy nową listę wierszy po podziale
new_rows = []
for _, row in original_df.iterrows():
    new_rows.extend(split_row(row.values, target_cols))

# Tworzymy nowy DataFrame
reshaped_df = pd.DataFrame(new_rows)

# Usuwamy nadmiarowe kolumny, jeśli istnieją (ostatni wiersz może być krótszy)
reshaped_df = reshaped_df.iloc[:, :target_cols]

print("Nowy wymiar DataFrame:", reshaped_df.shape)
```
