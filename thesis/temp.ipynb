```python
import matplotlib.pyplot as plt
import numpy as np

# Etykiety osi
labels = ['Pozytywna', 'Negatywna']

# Macierz z podpisami
conf_matrix_labels = np.array([['TP', 'FN'],
                               ['FP', 'TN']])

# Kolory – wyróżniamy TP i TN
colors = np.array([['#a8e6a1', '#e0e0e0'],   # TP (zielony), FN (szary)
                   ['#e0e0e0', '#f5a3a3']])  # FP (szary), TN (niebieski)

fig, ax = plt.subplots()

# Rysowanie kolorów jako tła
for i in range(2):
    for j in range(2):
        ax.add_patch(plt.Rectangle((j - 0.5, i - 0.5), 1, 1,
                                   color=colors[i, j]))

# Dodanie tekstu do komórek
for i in range(2):
    for j in range(2):
        ax.text(j, i, conf_matrix_labels[i, j],
                ha="center", va="center", color="black", fontsize=14, fontweight='bold')

# Ustawienia siatki i osi
ax.set_xticks(np.arange(2))
ax.set_yticks(np.arange(2))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
ax.set_xlim(-0.5, 1.5)
ax.set_ylim(1.5, -0.5)  # klasyczne odwrócenie osi Y

ax.set_xlabel('Etykieta przewidywana')
ax.set_ylabel('Etykieta rzeczywista')

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(False)
plt.show()
```
