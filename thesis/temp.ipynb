```python
import matplotlib.pyplot as plt
import numpy as np

# Etykiety osi
labels = ['Pozytywna', 'Negatywna']

# Macierz z podpisami
conf_matrix_labels = np.array([['TP', 'FN'],
                               ['FP', 'TN']])

# Kolory – wyróżniamy TP i TN
colors = np.array([['#a8e6a1', '#e0e0e0'],   # TP (zielony), FN (szary)
                   ['#e0e0e0', '#f5a3a3']])  # FP (szary), TN (niebieski)

fig, ax = plt.subplots()

# Rysowanie kolorów jako tła
for i in range(2):
    for j in range(2):
        ax.add_patch(plt.Rectangle((j - 0.5, i - 0.5), 1, 1,
                                   color=colors[i, j]))

# Dodanie tekstu do komórek
for i in range(2):
    for j in range(2):
        ax.text(j, i, conf_matrix_labels[i, j],
                ha="center", va="center", color="black", fontsize=14, fontweight='bold')

# Ustawienia siatki i osi
ax.set_xticks(np.arange(2))
ax.set_yticks(np.arange(2))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
ax.set_xlim(-0.5, 1.5)
ax.set_ylim(1.5, -0.5)  # klasyczne odwrócenie osi Y

ax.set_xlabel('Etykieta przewidywana')
ax.set_ylabel('Etykieta rzeczywista')

plt.gca().set_aspect('equal', adjustable='box')
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Przykładowe dane
np.random.seed(0)
n = 100

# Klasa 1
pred_a_class1 = np.random.beta(2, 5, n)
pred_b_class1 = pred_a_class1 * 0.5 + np.random.beta(2, 5, n) * 0.3

# Klasa 2
pred_a_class2 = np.random.beta(5, 2, n)
pred_b_class2 = pred_a_class2 * 0.5 + np.random.beta(5, 2, n) * 0.3

# Tworzymy DataFrame
df = pd.DataFrame({
    'Predictor A': np.concatenate([pred_a_class1, pred_a_class2]),
    'Predictor B': np.concatenate([pred_b_class1, pred_b_class2]),
    'Class': ['Class 1'] * n + ['Class 2'] * n
})

# Ustawienia kolorów dla klas
palette = {'Class 1': '#F88379', 'Class 2': '#9999FF'}

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Wykres rozrzutu Predictor A vs Predictor B
sns.scatterplot(data=df, x='Predictor A', y='Predictor B', hue='Class', style='Class', palette=palette, ax=axs[0])

axs[0].set_title('Scatter plot Predictor A vs Predictor B')
axs[0].legend(loc='upper left')

# Wykres gęstości dla Predictor A i Predictor B - oddzielne osie na prawym wykresie
# Używamy dwóch osi y na tym samym wykresie
ax1 = axs[1]
ax2 = ax1.twinx()

# Gęstość Predictor A dla obu klas
sns.kdeplot(data=df, x='Predictor A', hue='Class', palette=palette, ax=ax1)
ax1.set_ylabel('Density Predictor A')

# # Gęstość Predictor B dla obu klas
# sns.kdeplot(data=df, x='Predictor B', hue='Class', palette=palette, ax=ax2, linestyle='--')
# ax2.set_ylabel('Density Predictor B')

ax1.set_title('Density plots for Predictor A and B')
ax1.set_xlabel('Predictor Value')

# Legenda: scalamy legendy obu wykresów gęstości
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

# Usuwamy duplikaty w legendzie
by_label = dict(zip(labels1 + labels2, lines1 + lines2))
ax1.legend(by_label.values(), by_label.keys())

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Dane
np.random.seed(0)
n = 100

pred_a_class1 = np.random.beta(2, 5, n)
pred_b_class1 = pred_a_class1 * 0.5 + np.random.beta(2, 5, n) * 0.3

pred_a_class2 = np.random.beta(5, 2, n)
pred_b_class2 = pred_a_class2 * 0.5 + np.random.beta(5, 2, n) * 0.3

df = pd.DataFrame({
    'Predictor A': np.concatenate([pred_a_class1, pred_a_class2]),
    'Predictor B': np.concatenate([pred_b_class1, pred_b_class2]),
    'Class': ['Klasa A'] * n + ['Klasa B'] * n
})

palette = {'Klasa A': '#F88379', 'Klasa B': '#9999FF'}

# # 1. Scatter plot Predictor A vs Predictor B
plt.figure(figsize=(6, 5))
sns.scatterplot(data=df, x='Predictor A', y='Predictor B', hue='Class', style='Class', palette=palette)
plt.xlabel("Cecha A")
plt.ylabel("Cecha B")
plt.legend(title='Klasa', loc='upper left')
plt.tight_layout()
plt.show()

from matplotlib.lines import Line2D
fig, ax1 = plt.subplots(figsize=(6, 5))
sns.kdeplot(data=df, x='Predictor A', hue='Class', palette=palette, ax=ax1, legend=False)
ax1.set_xlabel('Wartość cechy A')
ax1.set_ylabel('Gęstość prawdopodobieństwa cechy A')
ax1.axvline(x=0.35, ymin=0, ymax=0.6255, color='black', linestyle='-', linewidth=1)
ax1.annotate(
    '',                      # pusty tekst (tylko strzałka)
    xy=(0.355, 0.93),            # początek strzałki (ogon)
    xytext=(-0.25, 0.93),    # koniec strzałki (głowa) — po lewej
    arrowprops=dict(
        arrowstyle='<-',     # strzałka skierowana w lewo
        color='#F88379',
        linewidth=1
    )
)

ax1.annotate(
    '',                      # pusty tekst (tylko strzałka)
    xy=(0.355, 0.135),            # początek strzałki (ogon)
    xytext=(-0.25, 0.135),    # koniec strzałki (głowa) — po lewej
    arrowprops=dict(
        arrowstyle='<-',     # strzałka skierowana w lewo
        color='#9999FF',
        linewidth=1
    )
)

# Tworzymy legendę z linii o kolorach jak na wykresie
lines = [Line2D([0], [0], color=palette[cls], lw=2) for cls in palette]
ax1.legend(handles=lines, labels=list(palette.keys()), title='Klasa', loc='upper right')

plt.tight_layout()
plt.show()


# # 3. Wykres gęstości Predictor B obrócony o 90 stopni
plt.figure(figsize=(6, 5))
sns.kdeplot(data=df, y='Predictor B', hue='Class', palette=palette)
plt.ylabel('Wartość cechy B')
plt.xlabel('Gęstość prawdopodobieństwa cechy B')
plt.tight_layout()

plt.axhline(y=0.35, xmin=0, xmax=0.355, color='black', linestyle='-', linewidth=1)



plt.annotate(
    '',                      # pusty tekst (tylko strzałka)
    xy=(0.15, 0.35),            # początek strzałki (ogon)
    xytext=(0.15, -0.096),    # koniec strzałki (głowa) — po lewej
    arrowprops=dict(
        arrowstyle='<-',     # strzałka skierowana w lewo
        color='#9999FF',
        linewidth=1
    )
)

plt.annotate(
    '',                      # pusty tekst (tylko strzałka)
    xy=(0.83, 0.35),            # początek strzałki (ogon)
    xytext=(0.83, -0.096),    # koniec strzałki (głowa) — po lewej
    arrowprops=dict(
        arrowstyle='<-',     # strzałka skierowana w lewo
        color='#F88379',
        linewidth=1
    )
)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_relu_sigmoid():
    # Zakres wartości wejściowych
    x = np.linspace(-10, 10, 400)

    # Funkcja ReLU
    relu = np.maximum(0, x)

    # Funkcja Sigmoid
    sigmoid = 1 / (1 + np.exp(-x))

    # Wykres ReLU
    plt.figure()
    plt.plot(x, relu, linewidth=2)
    # plt.title('Funkcja ReLU')
    plt.xlabel('x')
    plt.ylabel('ReLU(x)')
    # plt.grid(True)

    # Wykres Sigmoid
    plt.figure()
    plt.plot(x, sigmoid, linewidth=2)
    # plt.title('Funkcja Sigmoid')
    plt.xlabel('x')
    plt.ylabel('Sigmoid(x)')
    plt.grid(visible=False)
    

    plt.show()

# Uruchomienie funkcji
plot_relu_sigmoid()
```
